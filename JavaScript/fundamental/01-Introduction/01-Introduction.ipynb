# 📘 Chapter 01 - Introduction

## What is JavaScript?

JavaScript (JS) is a high-level, dynamically typed programming language used to make web pages interactive.

---

## History

- Created by **Brendan Eich**
- Year: **1995**
- Company: **Netscape**
- Standard: **ECMAScript (ES)**

---

## Features

- High-Level
- Dynamic
- Lightweight
- Object-Oriented
- Event-Driven
- Cross-Platform
- Interpreted (JIT)

---

## Uses

- Web Development
- Backend (Node.js)
- Mobile Apps (React Native)
- Desktop Apps (Electron)
- Games
- APIs

---

## JavaScript Engines

| Browser | Engine |

| Chrome | V8 |
| Edge | V8 |

| Safari | JavaScriptCore |

## JavaScript vs Java

| JavaScript | Java |
|------------|------|
| Dynamic | Static |
| Browser + Node.js | JVM |
| Web Development | Enterprise Applications |
