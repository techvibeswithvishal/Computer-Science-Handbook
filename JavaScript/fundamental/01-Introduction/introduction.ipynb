# 📖 What is JavaScript?

**JavaScript (JS)** is a high-level, lightweight, interpreted (or Just-In-Time compiled) programming language used to make websites interactive and dynamic.

Without JavaScript, websites are static.

JavaScript allows us to create:

- Interactive websites
- Dynamic web pages
- Web applications
- Mobile applications
- Desktop applications
- Backend servers
- Games

# ❌ Java vs JavaScript

Many beginners think Java and JavaScript are the same.

They are completely different languages.

| Java | JavaScript |
|------|------------|
| Compiled Language | Interpreted / JIT Compiled |
| Used for Enterprise Applications | Used for Web Development |
| Strongly Typed | Dynamically Typed |
| Runs on JVM | Runs in Browser & Node.js |

# ⚙️ How JavaScript Works

```
User

↓

Browser

↓

JavaScript Engine

↓

Execution

↓

Output
```

| Browser | Engine |

# 🛠 Ways to Run JavaScript

## 1. Browser Console

Press:

```
F12
```

Go to:

```
Console
```

Type:

```javascript
console.log("Hello World");
```

---

## 2. HTML File

```html
<!DOCTYPE html>
<html>

<head>
    <title>JavaScript</title>
</head>

<body>

<script>

console.log("Hello World");

</script>

</body>

</html>
```

---

## 3. External JavaScript File

Create

```
app.js
```

```javascript
console.log("Hello World");
```

Connect it:

```html
<script src="app.js"></script>
```

---

## 4. Node.js

Terminal

```bash
node app.js
```

# 📝 Comments in JavaScript

Comments are ignored by the JavaScript engine.

They are used to explain code.

---

## Single Line Comment

```javascript
// This is a comment
```

---

## Multi Line Comment

```javascript
/*
This is
a multi-line
comment
*/
```

---

# 📌 Best Practices

✅ Use meaningful variable names.

✅ Write comments only when necessary.

✅ Keep code readable.

✅ Use `let` and `const` instead of `var`.

---

# ⚠ Common Beginner Mistakes

❌ Thinking JavaScript = Java

❌ Forgetting to link the JavaScript file.

❌ Writing code outside the `<script>` tag.

❌ Using `var` everywhere.
