# 📘 Chapter 04 - Data Types

---

# 🎯 Goal

Learn different types of data JavaScript can store.

---

# 📌 What is a Data Type?

A **Data Type** defines the type of value stored in a variable.

Example:

```javascript
let age = 21;
```

Here, `21` is a **Number**.

---

# 📊 JavaScript Data Types

| Category | Data Types |
|----------|------------|
| Primitive | String, Number, Boolean, Undefined, Null, BigInt, Symbol |
| Non-Primitive (Reference) | Object, Array, Function |

---

# 🔹 Primitive Data Types

| Type | Description | Example |
|------|-------------|---------|
| String | Text | `"Vishal"` |
| Number | Integer & Decimal | `10`, `99.5` |
| Boolean | True / False | `true` |
| Undefined | Variable declared but not assigned | `let x;` |
| Null | Intentional empty value | `null` |
| BigInt | Large integers | `123n` |
| Symbol | Unique identifier | `Symbol("id")` |

---

# 🔹 Non-Primitive Data Types

| Type | Description |
|------|-------------|
| Object | Stores key-value pairs |
| Array | Stores multiple values |
| Function | Reusable block of code |

---

# 📌 Primitive vs Non-Primitive

| Primitive | Non-Primitive |
|------------|---------------|
| Stored by Value | Stored by Reference |
| Immutable | Mutable |
| Fixed Value | Can be Modified |

---

# 🧪 typeof Operator

Used to check the data type.

| Expression | Output |
|------------|--------|
| `"Hello"` | string |
| `10` | number |
| `true` | boolean |
| `undefined` | undefined |
| `123n` | bigint |
| `Symbol()` | symbol |
| `{}` | object |
| `[]` | object |
| `function(){}` | function |
| `null` | **object ⚠️** |

---

# ⚠ Interview Trap

```text
typeof null
```

Output:

```text
object
```

Reason:

- Historical bug in JavaScript.
- Kept for backward compatibility.

⭐ Frequently asked in interviews.

---

# 📌 Memory

| Type | Memory |
|------|--------|
| Primitive | Stack |
| Non-Primitive | Heap |

---

# 📌 Mutable vs Immutable

| Mutable | Immutable |
|----------|-----------|
| Object | String |
