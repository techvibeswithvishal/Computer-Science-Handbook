# 📘 Chapter 03 - Variables

---

# 🎯 Goal

Learn how to store data in JavaScript.

---

# 📌 What is a Variable?

A **variable** is a named container used to store data.

Example:

```
Name → Vishal
Age → 21
```

---

# 🧠 Variable Declaration

| Keyword | Can Reassign | Can Redeclare | Scope |
|----------|-------------|---------------|-------|
| var | ✅ Yes | ✅ Yes | Function |
| let | ✅ Yes | ❌ No | Block |
| const | ❌ No | ❌ No | Block |

---

# 📌 Syntax

| Keyword | Syntax |
|----------|--------|
| var | `var age = 20;` |
| let | `let age = 20;` |
| const | `const age = 20;` |

---

# 📊 var vs let vs const

| Feature | var | let | const |
|----------|-----|-----|-------|
| Scope | Function | Block | Block |
| Redeclare | ✅ | ❌ | ❌ |
| Reassign | ✅ | ✅ | ❌ |
| Hoisted | ✅ | ✅ (TDZ) | ✅ (TDZ) |
| Modern Use | ❌ | ✅ | ⭐ Best |

---

# 📌 Scope

| Scope | Accessible In |
|---------|--------------|
| Global | Everywhere |
| Function | Inside Function |
| Block | Inside `{ }` |

---

# 🌍 Variable Naming Rules

✅ Can contain:

- Letters
- Digits
- `_`
- `$`

---

❌ Cannot

- Start with number

```
1name ❌
```

---

- Use spaces

```
first name ❌
```

---

- Use reserved keywords

```
let let = 10 ❌
```

---

# ✅ Valid Names

```
userName
age
student1
_price
$user
```

---

# ❌ Invalid Names

```
1name
first name
let
class
```

---

# 📌 Best Naming Convention

| Style | Example |
|--------|---------|
| camelCase | firstName |
| PascalCase | StudentData |
| snake_case | first_name |
| UPPER_CASE | PI |

---

# 📌 Hoisting (Interview)

| Keyword | Hoisted | Initialized |
|----------|----------|-------------|
| var | ✅ | ✅ undefined |
| let | ✅ | ❌ TDZ |
| const | ✅ | ❌ TDZ |

---

# 📌 TDZ (Temporal Dead Zone)

Time between:

Variable declaration

↓

Initialization

Accessing variable before initialization causes:

```
ReferenceError
```

---

# 📌 const with Objects

```
const person = {};
```

✅ Object properties can change.

❌ Variable cannot be reassigned.

---

# 📌 Best Practices

✅ Prefer `const`

↓

If value changes

↓

Use `let`

↓

Avoid

```
var
```

---

# ⚠ Common Mistakes

❌ Using var everywhere

❌ Accessing let before declaration

❌ Reassigning const

❌ Poor variable names
