# 📘 Chapter 05 - Type Conversion

---

# 🎯 Goal

Learn how JavaScript converts one data type into another.

---

# 📌 What is Type Conversion?

**Type Conversion** is the process of converting one data type into another.

Example:

```
"100" → 100
100 → "100"
1 → true
```

---

# 📊 Types of Conversion

| Type | Description |
|------|-------------|
| Implicit Conversion | Automatically by JavaScript |
| Explicit Conversion | Manually by Programmer |

---

# 🔹 Implicit Conversion (Type Coercion)

JavaScript automatically converts data types when needed.

| Expression | Result |
|-----------|--------|
| `"5" + 2` | `"52"` |
| `"5" - 2` | `3` |
| `"5" * 2` | `10` |
| `"10" / 2` | `5` |
| `true + 1` | `2` |
| `false + 1` | `1` |
| `null + 1` | `1` |

---

# 🔹 Explicit Conversion

Programmer converts the data type manually.

| Function | Converts To |
|----------|-------------|
| Number() | Number |
| String() | String |
| Boolean() | Boolean |
| parseInt() | Integer |
| parseFloat() | Decimal |

---

# 📌 Number()

| Input | Output |
|-------|--------|
| `"100"` | `100` |
| `"10.5"` | `10.5` |
| `true` | `1` |
| `false` | `0` |
| `null` | `0` |
| `undefined` | `NaN` |
| `"Hello"` | `NaN` |

---

# 📌 String()

| Input | Output |
|-------|--------|
| `100` | `"100"` |
| `true` | `"true"` |
| `false` | `"false"` |
| `null` | `"null"` |
| `undefined` | `"undefined"` |

---

# 📌 Boolean()

### Falsy Values

| Value |
|-------|
| false |
| 0 |
| -0 |
| "" |
| null |
| undefined |
| NaN |

Everything else is **Truthy**.

---

# 📌 parseInt()

Converts a value into an integer.

| Input | Output |
|-------|--------|
| `"100"` | `100` |
| `"100px"` | `100` |
| `"10.5"` | `10` |
| `"Hello"` | `NaN` |

---

# 📌 parseFloat()

Converts a value into a decimal number.

| Input | Output |
|-------|--------|
| `"10.5"` | `10.5` |
| `"100"` | `100` |
| `"10.55px"` | `10.55` |

---

# 📌 Number() vs parseInt()

| Number() | parseInt() |
|-----------|------------|
| Converts entire value | Reads integer only |
| `"100px"` → NaN | `"100px"` → 100 |
| Decimal Supported | Decimal Removed |

---

# 📌 NaN (Not a Number)

NaN means **Invalid Number**.

Examples:

```
Number("Hello")

0 / 0

parseInt("ABC")
```

---

# 📌 Check NaN

Use

```
isNaN()
```

Returns:

- true
- false

---

# 📌 Truthy vs Falsy

## Truthy

```
1
-1
"Hello"
[]
{}
true
Infinity
```

---

## Falsy

```
false
0
-0
""
null
undefined
NaN
```

# ⚠ Common Interview Traps

| Expression | Result |
|-----------|--------|
| `"5" + 5` | `"55"` |
| `"5" - 5` | `0` |
| `true + true` | `2` |
| `false + true` | `1` |
| `null == undefined` | true |
| `null === undefined` | false |

---

# 📌 Best Practices

✅ Use `Number()` for numeric conversion.

✅ Use `parseInt()` only for integers.

✅ Validate input before conversion.

✅ Use `===` instead of `==`.
