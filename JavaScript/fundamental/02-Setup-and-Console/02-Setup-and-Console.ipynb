# 📘 Chapter 02 - Setup & Console

---

# 🎯 Goal

Learn how to run JavaScript and debug programs.

---

# 🛠 Required Software

| Software | Purpose |
|----------|---------|
| VS Code | Code Editor |
| Chrome | Run JavaScript |
| Node.js | Run JS outside Browser |
| Live Server | Auto Refresh |

---

# 🚀 Ways to Run JavaScript

| Method | Description | File Required |
|---------|-------------|--------------|
| Browser Console | Test code quickly | ❌ |
| Internal JS | Inside HTML | index.html |
| External JS | Separate JS file | app.js |
| Node.js | Terminal execution | app.js |

---

# 📂 Internal vs External JavaScript

| Internal | External |
|----------|----------|
| Inside HTML | Separate JS File |
| Small Programs | Large Projects |
| Not Reusable | Reusable |
| Difficult to Maintain | Easy to Maintain |

---

# 📋 Console Methods

| Method | Purpose |
|---------|---------|
| console.log() | Normal Output |
| console.error() | Error Message |
| console.warn() | Warning |
| console.info() | Information |
| console.table() | Display Table |
| console.clear() | Clear Console |
| console.time() | Start Timer |
| console.timeEnd() | Stop Timer |

---

# 🌐 Browser Developer Tools

| Shortcut | Action |
|----------|--------|
| F12 | Open DevTools |
| Ctrl + Shift + I | Open DevTools |
| Ctrl + Shift + J | Open Console |

---

# 📑 Important DevTools Tabs

| Tab | Use |

# 📁 Live Server

# ⚡ Node.js Commands

| Command | Description |
|----------|-------------|
| node -v | Check Version |
| npm -v | Check npm Version |
| node app.js | Run JavaScript |

---

# 📌 Best Practices

✅ Use External JavaScript

✅ Keep `<script>` before `</body>`

✅ Use meaningful console messages

✅ Remove console logs before production

---

# ❌ Common Mistakes

| Wrong | Correct |
|--------|---------|
| Wrong src path | Correct src path |
| JS in wrong place | JS before `</body>` |
| Expect `console.log()` on webpage | Check Console |
| Forget Save | Save before Refresh |
