# 📖 Ways to Run JavaScript

JavaScript can run in several ways.

## Method 1: Browser Console

Open Developer Tools.

```
F12
```

Select:

```
Console
```

Type:

```javascript
console.log("Hello World");
```

Output:

```
Hello World
```

---

## Method 2: Internal JavaScript

```html
<!DOCTYPE html>
<html>
<body>

<script>
console.log("Hello World");
</script>

</body>
</html>
```

---

## Method 3: External JavaScript

Create:

```
app.js
```

```javascript
console.log("Hello World");
```

Connect it:

```html
<script src="app.js"></script>
```

---

## Method 4: Node.js

Terminal:

```bash
node app.js
```

Output:

```
Hello World
```

# 📖 The Console Object

The `console` object is used to display information, debug programs, and inspect values.

---

# Console Methods

## console.log()

Displays normal messages.

```javascript
console.log("Hello JavaScript");
```

---

## console.error()

Displays an error message.

```javascript
console.error("Something went wrong.");
```

---

## console.warn()

Displays a warning.

```javascript
console.warn("This is a warning.");
```

---

## console.info()

Displays informational messages.

```javascript
console.info("Information");
```

---

## console.clear()

Clears the console.

```javascript
console.clear();
```

---

## console.table()

Displays data in table format.

```javascript
const student = {
    name: "Vishal",
    age: 21,
    course: "B.Tech"
};

console.table(student);
```

---

## console.time()

Starts a timer.

```javascript
console.time("Timer");

console.log("Learning JavaScript");

console.timeEnd("Timer");
```

---

# Debugging

Debugging means finding and fixing errors in a program.

The console is one of the most important debugging tools.

Example:

```javascript
let age = 20;

console.log(age);
```

---

# Best Practices

✅ Use `console.log()` while learning.

✅ Remove unnecessary console statements before publishing a project.

✅ Use `console.error()` for errors.

✅ Use `console.table()` for arrays and objects.

---

# Common Beginner Mistakes

❌ Forgetting to save the file.

❌ Incorrect file path in `src`.

```html
<script src="app.js"></script>
```

❌ Writing JavaScript outside the `<script>` tag.

❌ Expecting `console.log()` output to appear on the webpage.
