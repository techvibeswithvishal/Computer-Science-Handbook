# 📘 Chapter 04 - Styling

---

# 🎯 Goal

Learn how styling works in React Native and how it differs from CSS.

---

# 📌 What is Styling?

**Styling** is used to design the User Interface (UI) of an application.

React Native uses **JavaScript objects** instead of CSS files.

# 📊 CSS vs React Native

| CSS | React Native |
|-----|--------------|
| `.css` File | JavaScript Object |
| class | style |
| px Unit | No Unit Required |
| kebab-case | camelCase |
| margin-top | marginTop |

# ⚠ Common Mistakes

| Wrong | Correct |
|--------|---------|
| CSS File | StyleSheet |
| margin-top | marginTop |
| width: "100px" | width: 100 |
| class | style |
