# 📊 Required Software

| Software | Purpose |
|----------|---------|

| JDK (Java 17+) | Android Build |
| Git | Version Control |

---

# 📌 Recommended Framework

| Framework | Recommendation |
|-----------|---------------|
| Expo | ⭐ Beginner |
| React Native CLI | ⭐ Advanced |

---

# 📊 Expo vs React Native CLI

| Expo | React Native CLI |
|------|------------------|
| Easy Setup | Manual Setup |
| Beginner Friendly | Advanced |
| Limited Native Modules | Full Native Access |
| Fast Development | Better Customization |
| OTA Updates | Native Build Required |

---

# 📌 Create Expo Project

| Command | Purpose |
|----------|---------|
| `npx create-expo-app app-name` | Create Project |
| `cd app-name` | Open Project |
| `npm start` | Start Server |

---

# 📌 React Native CLI

| Command | Purpose |
|----------|---------|
| `npx react-native init AppName` | Create Project |
| `npx react-native run-android` | Run Android |
| `npx react-native run-ios` | Run iOS (macOS Only) |

---

# 📂 Project Structure

```text
project/
│
├── assets/
├── node_modules/
├── App.js
├── package.json
├── app.json
└── babel.config.js
```

# 📌 Important Files

| File | Purpose |
|------|---------|

# 📌 Development Tools

| Tool | Purpose |
|------|---------|

---

# 📌 React Native Build Flow

```text
Write Code
      │
      ▼
Metro Bundler
      │
      ▼
Expo / CLI
      │
      ▼
Android / iOS Device
```

---

# 📌 Metro Bundler

- JavaScript Bundler
- Compiles App
- Supports Hot Reload
- Fast Refresh

---

# 📌 Hot Reload vs Fast Refresh

| Hot Reload | Fast Refresh |
|------------|--------------|
| Updates UI | Updates UI + State |
| Older Method | Recommended |
| May Lose State | Preserves State |

---

# 📌 Best Practices

✅ Use Expo for learning

✅ Use React Native CLI for advanced apps

✅ Keep project structure organized

✅ Use Git from Day 1

---

# ⚠ Common Errors

| Error | Solution |
|-------|----------|
| npm not found | Install Node.js |
| Android SDK Missing | Install Android Studio |
| Emulator Not Running | Start Emulator |
| Module Not Found | Run `npm install` |
| Build Failed | Check JDK & SDK |
