---

# 🎯 Goal

Learn the basic building blocks of a React Native application.

---

# 📌 What is a Component?

A **Component** is a reusable piece of UI.

Everything visible on the screen is a component.

---

# 📊 Component Hierarchy

```text
App
│
├── Header
├── Profile
├── Button
├── Card
└── Footer
```

---

# 📊 Types of Components

| Type | Description |
|------|-------------|

# 📊 Web vs React Native Components

| React (Web) | React Native |

| p | Text |
| img | Image |

# 📌 Best Practices

✅ Small Components

✅ Reusable Components

✅ PascalCase Naming

✅ One Component → One File

✅ Keep Components Independent

---

# ⚠ Common Mistakes

| Wrong | Correct |
|--------|---------|
| component.jsx | Component.jsx |
| Large Components | Small Components |
| Duplicate UI | Reusable Components |
| Inline Everything | Separate Components |
