# рҹ“ҳ Chapter 05 - Flexbox

---

# рҹҺҜ Goal

Learn how to create responsive layouts using **Flexbox** in React Native.

---

# рҹ“Ң What is Flexbox?

**Flexbox** is the default layout system in React Native used to arrange components.

> React Native uses **Flexbox by default**.

---

# рҹ“Ҡ Default Values

| Property | Default |
|----------|---------|
| display | flex |
| flexDirection | column |
| justifyContent | flex-start |
| alignItems | stretch |

---

---

# рҹ“Ң flexDirection

| Value | Direction |
|-------|-----------|
| column | Top вҶ’ Bottom вӯҗ |
| row | Left вҶ’ Right |
| column-reverse | Bottom вҶ’ Top |
| row-reverse | Right вҶ’ Left |

---

# рҹ“Ң justifyContent

**Controls items on the Main Axis**

| Value | Description |
|-------|-------------|
| flex-start | Start |
| center | Center |
| flex-end | End |
| space-between | Equal Space Between |
| space-around | Equal Space Around |
| space-evenly | Equal Space Everywhere |

---

# рҹ“Ң alignItems

**Controls items on the Cross Axis**

| Value | Description |
|-------|-------------|
| stretch | Default |
| flex-start | Start |
| center | Center |
| flex-end | End |
| baseline | Text Baseline |

---

# рҹ“Ң flex

Controls available space.

| Value | Meaning |
|-------|---------|
| flex: 1 | Take Remaining Space |
| flex: 2 | Twice the Space |
| flex: 3 | Three Times the Space |

---

# рҹ“Ң alignSelf

Overrides `alignItems` for a single component.

| Value |
|-------|
| auto |
| flex-start |
| center |
| flex-end |
| stretch |

---

# рҹ“Ң flexWrap

Controls wrapping.

| Value | Description |
|-------|-------------|
| nowrap | Default |
| wrap | Wrap Items |
| wrap-reverse | Reverse Wrap |

---

# рҹ“Ң gap

Adds spacing between components.

| Property | Purpose |
|----------|---------|
| gap | All Directions |
| rowGap | Vertical Gap |
| columnGap | Horizontal Gap |

---

# рҹ“Ң Common Layouts

| Layout | flexDirection |
|---------|---------------|
| Vertical List | column |
| Horizontal List | row |
| Card Layout | column |
| Navigation Bar | row |

---

# рҹ“Ҡ justifyContent vs alignItems

| justifyContent | alignItems |
|---------------|------------|
| Main Axis | Cross Axis |
| Horizontal (row) | Vertical (row) |
| Vertical (column) | Horizontal (column) |

---

# рҹ“Ң Best Practices

вң… Use `flex:1` for screens

вң… Use `justifyContent` for Main Axis

вң… Use `alignItems` for Cross Axis

вң… Prefer Flexbox over Absolute Positioning

---

# вҡ  Common Mistakes

| Wrong | Correct |
|--------|---------|
| Assuming row is default | column is default |
| Confusing Main/Cross Axis | Learn Axis First |
| Overusing absolute | Use Flexbox |
