# 📘 Chapter 01 - Introduction to React Native

---

# 🎯 Goal

Understand what React Native is, why it is used, and how it differs from React and native app development.

---

# 📌 What is React Native?

**React Native** is an **open-source framework** used to build **cross-platform mobile applications** using **JavaScript** and **React**.

- Developed by **Meta (Facebook)**
- Released in **2015**
- Uses React Concepts
- Builds Native Mobile Apps

---

# 📊 React Native at a Glance

| Property | Description |
|----------|-------------|
| Type | Mobile App Framework |
| Developed By | Meta (Facebook) |
| Initial Release | 2015 |
| Language | JavaScript |
| Based On | React |
| Platforms | Android & iOS |

---

# 📌 Why React Native?

- Cross-Platform Development
- Single Codebase
- Native Performance
- Reusable Components
- Faster Development
- Huge Community
- Hot Reloading

---

# 📊 React vs React Native

| React | React Native |
|--------|--------------|
| Web Applications | Mobile Applications |
| HTML | Native Components |
| CSS | StyleSheet API |
| Browser | Android & iOS |
| ReactDOM | Native Bridge |

---

# 📊 React Native vs Native Development

| React Native | Native Android/iOS |
|--------------|-------------------|
| One Codebase | Separate Codebases |
| Faster Development | More Development Time |
| JavaScript | Kotlin / Swift |
| Cross Platform | Platform Specific |
| Easier Maintenance | Higher Maintenance |

---

# 📌 React Native Architecture

```text
React Native App
        │
        ▼
JavaScript
        │
        ▼
React Native Bridge
        │
        ▼
Native Components
        │
   Android / iOS
```

# 📌 Advantages

- Cross Platform
- Native UI
- Fast Development
- Live Reload
- Large Community
- Reusable Code
- High Performance

---

# 📌 Limitations

- Large App Size
- Some Native Modules Required
- Performance slightly lower than Pure Native Apps
- Platform-specific code sometimes needed
