# 📘 Chapter 04 - Components

---

# 🎯 Goal

Understand React Components, their types, and why they are the foundation of React applications.

---

# 📌 What is a Component?

A **Component** is an independent, reusable piece of UI.

Think of a webpage as a collection of components.

Example:

```
Website
│
├── Navbar
├── Sidebar
├── Hero
├── Cards
├── Footer
```

---

# 📊 Component at a Glance

| Property | Description |
|----------|-------------|
| Type | JavaScript Function / Class |
| Purpose | Reusable UI |
| Returns | JSX |
| File Extension | `.jsx` (Recommended) |

---

# 📌 Why Components?

- Reusable
- Easy Maintenance
- Better Code Organization
- Independent
- Faster Development

---

# 📊 Types of Components

| Type | Status |
|------|--------|
| Functional Component | ✅ Recommended |
| Class Component | ⚠ Legacy |

---

# 📊 Functional vs Class Component

| Functional | Class |
|------------|-------|
| Function | ES6 Class |
| Hooks Supported | No Hooks |
| Less Code | More Code |
| Faster to Write | Complex |
| Recommended | Legacy |

---

# 📌 Component Naming Rules

| Rule | Example |
|------|---------|
| First Letter Capital | `Navbar` |
| PascalCase | `UserProfile` |
| One Component = One File | ✅ |

---

# 📌 Component Structure

```
Component
      │
      ▼
Return JSX
      │
      ▼
Render UI
```

---

# 📌 Reusability

One component can be used multiple times.

Example

```
Card

↓

Card

↓

Card

↓

Card
```

Write Once → Use Anywhere

---

# 📌 Component Hierarchy

```
App
│
├── Navbar
├── Hero
├── About
├── Services
├── Contact
└── Footer
```

# 📊 Parent vs Child Component

| Parent | Child |
|---------|-------|
| Contains Components | Nested Inside Parent |
| Passes Props | Receives Props |
