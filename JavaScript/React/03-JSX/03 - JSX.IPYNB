# 📘 Chapter 03 - JSX (JavaScript XML)

---

# 🎯 Goal

Understand what JSX is, why React uses it, and how it differs from HTML.

---

# 📌 What is JSX?

**JSX (JavaScript XML)** is a syntax extension for JavaScript that allows writing **HTML-like code inside JavaScript**.

- Easier UI Development
- More Readable
- Converted to JavaScript by Babel

---

# 📊 JSX at a Glance

| Property | Description |
|----------|-------------|
| Full Form | JavaScript XML |
| Type | JavaScript Syntax Extension |
| Compiled By | Babel |
| Used In | React Components |

---

# 📌 Why JSX?

- Easy to Read
- HTML-like Syntax
- Better Developer Experience
- Supports JavaScript Expressions
- Component Friendly

---

# 📊 JSX vs HTML

| HTML | JSX |
|------|-----|
| class | className |
| for | htmlFor |
| onclick | onClick |
| tabindex | tabIndex |
| style="" | style={{}} |

---

# 📌 JSX Rules

| Rule | Description |
|------|-------------|
| Single Parent | One parent element required |
| Closed Tags | Every tag must close |
| camelCase | Event names & attributes |
| Expressions | Use `{}` |
| Components | Start with Capital Letter |

---

# 📊 Self Closing Tags

| HTML | JSX |
|------|-----|
| `<br>` | `<br />` |
| `<img>` | `<img />` |
| `<hr>` | `<hr />` |
| `<input>` | `<input />` |

---

# 📌 JavaScript Inside JSX

Use **Curly Braces `{}`**

Can contain:

- Variables
- Functions
- Expressions
- Objects (limited)
- Calculations

Cannot contain:

- if...else
- for loop
- switch

Use:

- Ternary Operator
- map()
- Logical &&

---

# 📊 Expressions Allowed

| Allowed | Not Allowed |
|----------|-------------|
| Variables | if Statement |
| Function Calls | for Loop |
| Math Expressions | switch |
| Ternary | while |

---

# 📌 JSX Attributes

| HTML | JSX |
|------|------|
| class | className |
| for | htmlFor |
| onclick | onClick |
| readonly | readOnly |
| maxlength | maxLength |

---

# 📌 Comments in JSX

| Type | Syntax |
|------|---------|
| JavaScript | `{/* Comment */}` |

---

# 📌 Fragments

Used to avoid unnecessary wrapper elements.

Types:

- `<></>`
- `<React.Fragment></React.Fragment>`

---

# 📌 Babel

**Babel** converts JSX into regular JavaScript.

Flow:

```
JSX
   │
   ▼
Babel
   │
   ▼
JavaScript
   │
   ▼
Browser
```

# 📌 Best Practices

✅ Use Fragments when possible.

✅ Use `className` instead of `class`.

✅ Use camelCase attributes.

✅ Keep JSX clean and readable.
