# 📌 What is React?

**React** is an open-source JavaScript library for building **User Interfaces (UI)**, especially **Single Page Applications (SPA)**.

- Developed by **Meta (Facebook)**
- Released in **2013**
- Maintained by **Meta & Community**

---

# 📊 React at a Glance

| Feature | Description |
|---------|-------------|
| Type | JavaScript Library |
| Developed By | Meta (Facebook) |
| Initial Release | 2013 |
| Latest Architecture | Component-Based |
| Used For | Building User Interfaces |
| Platform | Web & React Native |

---

# 📌 Why React?

- Reusable Components
- Fast Rendering
- Virtual DOM
- Easy State Management
- Huge Community
- Rich Ecosystem
- High Industry Demand

---

# 📊 React vs Vanilla JavaScript

| React | JavaScript |
|--------|------------|
| Component-Based | Script-Based |
| Virtual DOM | Real DOM |
| Reusable UI | Repeated Code |
| Faster Updates | Manual DOM Manipulation |
| Large Applications | Small Projects |

---

# 📌 Core Concepts

| Concept | Description |
|----------|-------------|
| Components | Reusable UI Blocks |
| JSX | HTML-like Syntax |
| Props | Pass Data Between Components |
| State | Manage Dynamic Data |
| Hooks | Add Features to Functional Components |
| Virtual DOM | Efficient UI Updates |

---

# 📊 Real DOM vs Virtual DOM

| Real DOM | Virtual DOM |
|-----------|-------------|
| Updates Entire DOM | Updates Changed Part Only |
| Slower | Faster |
| Direct Manipulation | Virtual Copy |
| Less Efficient | More Efficient |
