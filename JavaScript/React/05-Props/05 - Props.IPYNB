# 📘 Chapter 05 - Props

---

# 🎯 Goal

Learn how to pass data from one component to another using **Props**.

---

# 📌 What are Props?

**Props (Properties)** are used to pass **data from Parent Component → Child Component**.

- Read-Only
- Immutable
- One-Way Data Flow

---

# 📊 Props at a Glance

| Property | Description |
|----------|-------------|
| Full Form | Properties |
| Purpose | Pass Data |
| Direction | Parent → Child |
| Mutable | ❌ No |
| Read Only | ✅ Yes |

---

# 📌 Why Props?

- Reusable Components
- Dynamic UI
- Pass Data
- Pass Functions
- Component Communication

---

# 📊 What Can Props Pass?

| Data Type | Supported |
|-----------|-----------|
| String | ✅ |
| Number | ✅ |
| Boolean | ✅ |
| Object | ✅ |
| Array | ✅ |
| Function | ✅ |
| JSX | ✅ |
| Component | ✅ |

---

# 📊 Parent vs Child

| Parent Component | Child Component |
|------------------|-----------------|
| Sends Props | Receives Props |
| Owns Data | Uses Data |
| Can Update Data | Cannot Modify Props |

---

# 📌 One-Way Data Flow

```
Parent
   │
   ▼
 Props
   │
   ▼
Child
```

React follows **One-Way Data Binding**.

---

# 📌 Props are Immutable

| Operation | Allowed |
|-----------|---------|
| Read Props | ✅ |
| Modify Props | ❌ |
| Pass New Props | ✅ (Parent) |

---

# 📊 Props vs State

| Props | State |
|-------|-------|
| Read Only | Mutable |
| Parent Controls | Component Controls |
| Parent → Child | Inside Component |
| Used for Communication | Used for Dynamic Data |

---

# 📌 Types of Props

- Simple Props
- Default Props
- Children Props
- Function Props
- Object Props
- Array Props

---

# 📌 children Prop

Used to pass UI/content between opening and closing component tags.

```
<Component>
    Content
</Component>
```

---

# 📌 Default Props

Provide default values when no prop is passed.

Used to avoid `undefined`.

---

# 📌 Destructuring Props

Instead of using:

```
props.name
```

Use:

```
{name}
```

Advantages:

- Cleaner Code
- Better Readability

---

# 📌 Prop Drilling

Passing props through multiple components.

```
App
 │
 ▼
Navbar
 │
 ▼
Menu
 │
 ▼
Button
```

Problem:

- Too many intermediate components.

Solution:

- Context API
- Redux Toolkit

---

# 📌 Best Practices

✅ Keep Props Read-Only

✅ Use Destructuring

✅ Use Meaningful Prop Names

✅ Avoid Deep Prop Drilling

---

# ⚠ Common Mistakes

| Wrong | Correct |
|--------|---------|
| Changing Props | Keep Props Immutable |
| Too Many Props | Group Related Data |
| Deep Prop Drilling | Use Context API |
