# 📌 Verify Installation

| Command | Purpose |
|---------|---------|
| `node -v` | Check Node.js Version |
| `npm -v` | Check npm Version |

---

# 📌 Create React Project (Vite)

| Command | Purpose |
|---------|---------|
| `npm create vite@latest` | Create React Project |
| `cd project-name` | Open Project |
| `npm install` | Install Dependencies |
| `npm run dev` | Start Development Server |

---

# 📊 React Project Structure

```text
project/
│
├── node_modules/
├── public/
├── src/
│   ├── assets/
│   ├── App.jsx
│   ├── main.jsx
│   └── index.css
│
├── package.json
├── package-lock.json
├── vite.config.js
└── index.html
```

---

| File | Purpose |

# 📌 Why Vite?

| Vite | Create React App |
|------|------------------|
| Faster | Slower |
| Instant Reload | Slow Reload |
| Lightweight | Heavy |
| Modern | Older Setup |

---

# ⚠ Common Errors

| Error | Solution |
|-------|----------|
| `'npm' not recognized` | Install Node.js |
| `Module not found` | Run `npm install` |
| Port already in use | Change Port / Close Process |
| Blank Screen | Check `main.jsx` & `App.jsx` |
| Missing node_modules | Run `npm install` |

---

# 📌 Best Practices

✅ Use Vite

✅ Keep Components inside `src/`

✅ Install only required packages

✅ Use Git from Day 1

# 📝 Remember

✔ Install Node.js before React.

✔ npm manages packages.

✔ Vite is the recommended React build tool.

✔ `main.jsx` is the entry point.

✔ `App.jsx` is the root component.

✔ `package.json` stores project information and dependencies.
