# 📘 09_Continue

## Learning Objectives

- Skip current iteration.

---

## What is Continue?

Skips the current iteration and moves to the next one.

---

## Syntax

```python
continue
```

---

## Example

```python
for i in range(6):
    if i == 3:
        continue
    print(i)
```
