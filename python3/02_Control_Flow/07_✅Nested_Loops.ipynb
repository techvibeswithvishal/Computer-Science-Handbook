## What are Nested Loops?

A loop inside another loop.

---

## Syntax

```python
for i in range(3):
    for j in range(3):
        print(i, j)
```

---

## Example

```python
for i in range(3):
    for j in range(3):
        print("*", end="")
    print()
```
