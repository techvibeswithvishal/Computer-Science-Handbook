## What is Nested If?

An `if` statement inside another `if` statement.

---

## Syntax

```python
if condition1:
    if condition2:
        statements
```

---

## Example

```python
age = 20
citizen = True

if age >= 18:
    if citizen:
        print("Eligible")
```
