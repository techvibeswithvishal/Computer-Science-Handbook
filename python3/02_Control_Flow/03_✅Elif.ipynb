## What is Elif?

`elif` is used to test multiple conditions after an `if`.

---

## Syntax

```python
if condition1:
    ...
elif condition2:
    ...
else:
    ...
```

---

## Example

```python
marks = 82

if marks >= 90:
    print("A")
elif marks >= 75:
    print("B")
else:
    print("C")
```
