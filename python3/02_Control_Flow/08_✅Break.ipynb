# 📘 08_Break

## Learning Objectives

- Stop loop execution.

---

## What is Break?

Terminates the nearest loop immediately.

---

## Syntax

```python
break
```

---

## Example

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```
