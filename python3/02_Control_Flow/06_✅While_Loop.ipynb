## What is While Loop?

Executes while the condition remains `True`.

---

## Syntax

```python
while condition:
    statements
```

---

## Example

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

## Summary

- Best for unknown iterations.

````python
x = int(input("Enter the number: "))
n = 1

while n <= 10:
    print(x * n)
    n += 1

    
````
