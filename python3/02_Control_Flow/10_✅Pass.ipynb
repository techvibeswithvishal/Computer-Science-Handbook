## Learning Objectives

---

## What is Pass?

`pass` is a null statement used when no action is required.

---

## Syntax

```python
pass
```

---

## Example

```python
if True:
    pass
```

---

## Practice Questions

1. Empty function.
2. Empty class.
3. Empty loop.
4. Empty if block.
5. Future implementation.

---

## Interview Questions

- Why use `pass`?
- Difference between `pass` and `continue`?

---

## Summary

- Placeholder statement.
- Prevents syntax errors when a block is intentionally left empty.

````python
for i in range(6):
    if i == 3:
        pass
    print(i)
````
