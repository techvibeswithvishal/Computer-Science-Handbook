# Keywords in Python

* Python is a case-sensitive programming language.
* Keywords are reserved words with special meanings.
* Keywords cannot be used as variable names, function names, or class names.
* Python provides built-in keywords to define the syntax and structure of programs.

### Find All Keywords

```python
import keyword

print(keyword.kwlist)
```

### Output

```python
['False', 'None', 'True', 'and', 'as', 'assert', 'async',
'await', 'break', 'class', 'continue', 'def', 'del',
'elif', 'else', 'except', 'finally', 'for', 'from',
'global', 'if', 'import', 'in', 'is', 'lambda',
'nonlocal', 'not', 'or', 'pass', 'raise', 'return',
'try', 'while', 'with', 'yield']
```

### Example

```python
if True:
    print("Hello Python")
```

Here, `if` and `True` are keywords.


# Identifiers in Python

Identifiers are names used for variables, functions, classes, modules, and other objects.

### Rules for Identifiers

* Must start with a letter (A-Z, a-z) or underscore (_).
* Can contain letters, digits (0-9), and underscore (_).
* Cannot start with a digit.
* Cannot be a Python keyword.
* Python is case-sensitive.

### Valid Identifiers

```python id="h2z3sk"
name
_age
student1
total_marks
```

### Invalid Identifiers

```python id="p2k6zc"
1name
class
student-name
total marks
```
