# Python Operators

Python has **7 types of operators**:


Trick **AB CLAIM**

1. **Arithmetic Operators** (`+`, `-`, `*`, `/`, `%`, `**`, `//`)
2. **Comparison Operators** (`==`, `!=`, `>`, `<`, `>=`, `<=`)
3. **Logical Operators** (`and`, `or`, `not`)
4. **Bitwise Operators** (`&`, `|`, `^`, `~`, `<<`, `>>`)
5. **Assignment Operators** (`=`, `+=`, `-=`, `*=`, `/=`, `%=`, `**=`, `//=`)
6. **Identity Operators** (`is`, `is not`)
7. **Membership Operators** (`in`, `not in`)


```python
a = 2
b = 4
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a**b)
print(a//b)
```

```python
a = 2
b = 4
print(a==b)
print(a!=b)
print(a<b)
print(a>b)
print(a>=b)
print(a<=b)
```

```python
a = True
b = False
print( a and b)
print( a or b)
print(not a)
```

```python
x = 2
y = 3
print(x & y) # 2
print(x | y) # 3
print(x ^ y) # 1
print(~x)    # -3
print(x << 2) # 8 
print(y >> 1) # 1
```

```python
a = 1

a += 1      # a = a + 1
print(a)    # 2

a -= 1      # a = a - 1
print(a)    # 1

a *= 2      # a = a * 2
print(a)    # 2

a /= 2      # a = a / 2
print(a)    # 1.0

a %= 2      # a = a % 2
print(a)    # 1.0

a **= 3     # a = a ** 3
print(a)    # 1.0

a //= 1     # a = a // 1
print(a)    # 1.0
```

```python
a = 10
b = 10

print(a is b)      # True
print(a is not b)  # False
```

```python
x = "Delhi"

print("D" in x)       # True
print("A" in x)       # False
print("D" not in x)   # False
```
