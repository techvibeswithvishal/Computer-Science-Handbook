# Literals in Python

Literals are fixed values assigned to variables.

### Types of Literals

### 1. Numeric Literals

Used to represent numbers.

```python
a = 100      # Integer
b = 10.5     # Float
c = 3 + 4j   # Complex
```

### 2. String Literals

Used to represent text.

```python
name = "Vishal"
city = 'Aligarh'
```

### 3. Boolean Literals

Represent logical values.

```python
is_student = True
is_admin = False
```

### 4. Special Literal

`None` represents the absence of a value.

```python
x = None
```


````python
print(True + True)
print(False + 10)
````

````python
print(True + 10)
print(0b1010)
print(type(3 + 4j))
print(bool(""))
print(None == False)
````
