## sorted()

```python
arr = [5,2,8,1]

print(sorted(arr))
```

Output:

```python
[1,2,5,8]
```

---

## enumerate()

```python
arr = ["a","b","c"]

for index,value in enumerate(arr):
    print(index,value)
```

Output:

```python
0 a
1 b
2 c
```

---

## zip()

```python
name = ["Vishal","Rahul"]
marks = [90,80]

for n,m in zip(name,marks):
    print(n,m)
```

Output:

```python
Vishal 90
Rahul 80
```

# Important DSA Modules

## collections

```python
from collections import deque
```

Queue Example:

```python
q = deque()

q.append(10)
q.append(20)

q.popleft()
```

Used in:
- BFS
- Queue

---

## heapq

```python
import heapq
```

Min Heap Example:

```python
heap = []

heapq.heappush(heap,10)
heapq.heappush(heap,5)

print(heapq.heappop(heap))
```

Used in:
- Priority Queue
- Dijkstra Algorithm
- Greedy Algorithms

# Loop Control Statements

## break

Stops loop immediately.

```python
for i in range(10):
    if i == 5:
        break
```

---

## continue

Skips current iteration.

```python
for i in range(10):
    if i == 5:
        continue
```

---

## pass

Does nothing.

```python
if True:
    pass
```
