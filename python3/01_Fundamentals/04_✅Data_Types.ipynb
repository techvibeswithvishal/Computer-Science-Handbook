# Python Data Types

Python data types can be divided into three categories:

### 1. Basic Data Types

`int`, `float`, `complex`, `bool`, `str`

### 2. Container Data Types

`list`, `tuple`, `set`, `dict`

### 3. User-Defined Data Types

`class`

Data types define the kind of value that can be stored and manipulated in a Python program.


```python
#int, float, complex, bool, str
print(4)

print(4.5)

print(4+4j)

print(True)
print(False)

print("hello sir")
```

```python
#list, tuple, set, dict
print([1, 2, 3])

print((1, 2, 3))

print({1, 2, 3})

print({"name": "Vishal", "age": 2000})
```

```python
## User-Defined Data Types

### Class (class)


class Student:
    pass

print(Student)


#second example


class Student:
    name = "Vishal"
    age = 2000

print(Student.name)
print(Student.age)
```
