1. Create
   

```python
name = "Vishal" 
print(name)
```

2. Access

```python
name = "Vishal" 
print(name[0]) 
print(name[-1]) 
```

3. Slicing

```python
print(name[0:3]) # Vis 
print(name[::-1]) # lahsiV
```

4. Add (Concatenation)

```python
first = "Vishal"
last = "Singh" 
print(first + " " + last)
```

5. Repetition + other important
   

```python
print("Hi " * 3)
```

```python
name = "Vishal"
print("V" in name) 
print("x" in name)
```

```python
#find
text = "banana" 
print(text.find("n"))
```

```python
#Reolace
text = "Hello Vishal"
print(text.replace("Vishal","Singh"))
```

6. Upper / Lower

```python
name = "Vishal" 
print(name.upper())
print(name.lower())
```

7. Split

```python
text = "Python Java C++"
print(text.split())
```

8. Join

```python
words = ["Python","Java","C++"] 
print("-".join(words))
```

9. Strip

```python
name = " Vishal " 
print(name.strip())
```
