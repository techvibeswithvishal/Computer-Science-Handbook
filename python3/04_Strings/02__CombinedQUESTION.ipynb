# 🚀 String Master Interview Challenge

## Problem Statement

Given a string `S`, write a program to analyze and process the string.

Your program should:

1. Find the length of the string.
2. Count vowels and consonants.
3. Count the total number of words.
4. Find the frequency of each character.
5. Find the most frequent character.
6. Find the first repeating character.
7. Find the first non-repeating character.
8. Find all duplicate characters.
9. Remove duplicate characters while preserving order.
10. Reverse the complete string.
11. Reverse each word individually.
12. Reverse the order of words.
13. Check whether the entire string is a palindrome.
14. Find all palindrome words present in the string.
15. Find the longest word.
16. Find the shortest word.
17. Check whether two given words are anagrams.
18. Check whether all characters in the string are unique.
19. Check whether one string is a rotation of another.
20. Display the final results in a structured format.

### Input Example

```text
Madam Arora teaches malayalam and Java Programming
```

### Constraints

* String may contain uppercase and lowercase letters.
* String may contain spaces.
* Ignore case when performing palindrome and anagram checks.
* Do not use built-in reverse functions.

### Objective

Design an efficient solution using appropriate data structures and algorithms while analyzing the time and space complexity of each operation.

````python
#Count vowels and consonants.
s="Madam Arora teaches malayalam and Java Programming"

# count vowel

vowel = 0
for v in s:
     if v.lower() in "aeiou":
        vowel += 1

print("total vowel : " ,vowel)
     




#count constant

constant = 0
for c in s:
    if c.lower() in "bcdfghjklmnpqrstvwxyz":
       constant += 1

print("total constant :" ,constant)






````

````python
# Count the total number of words.
s="Madam Arora teaches malayalam and Java Programming"

words = 0

for i in s:
   if i == " ":
       words += 1

n = words + 1
print("total words : " , n)


# OR _ _
s = " Madam Arora teaches malayalam and Java Programming "

count = 0
new_word = True

for ch in s:

    if ch == " ":
        new_word = True

    elif new_word:
        count += 1
        new_word = False

print("total words : "  ,count)


#  OR
s = " Madam Arora teaches malayalam and Java Programming "
words= len(s.split())
print("total words : " ,words)

````
