# 📖 What is a Subquery?

A **Subquery** (Nested Query) is a query written inside another SQL query.

The inner query executes first and its result is used by the outer query.

---

# Basic Syntax

```sql
SELECT column_name
FROM table_name
WHERE column_name OPERATOR
(
    SELECT column_name
    FROM another_table
);
```

# 1. Subquery with WHERE

Find employees whose salary is greater than the average salary.

```sql
SELECT *
FROM employees
WHERE salary >
(
    SELECT AVG(salary)
    FROM employees
);
```

---

# 2. Subquery with IN

Find employees working in departments that have employees with salary greater than 50000.

```sql
SELECT *
FROM employees
WHERE department IN
(
    SELECT department
    FROM employees
    WHERE salary > 50000
);
```

---

# 3. Subquery with EXISTS

Returns TRUE if the subquery returns at least one row.

```sql
SELECT *
FROM employees e
WHERE EXISTS
(
    SELECT *
    FROM employees
    WHERE department = e.department
);
```

---

# 4. Subquery with FROM

Use the result of one query as a temporary table.

```sql
SELECT AVG(salary)
FROM
(
    SELECT salary
    FROM employees
) AS temp;
```

---

# 5. Subquery with SELECT

Display employee name along with the average salary.

```sql
SELECT
name,
(
    SELECT AVG(salary)
    FROM employees
) AS Average_Salary
FROM employees;
```

---

# ANY Operator

Returns TRUE if the condition matches **at least one** value.

```sql
SELECT *
FROM employees
WHERE salary > ANY
(
    SELECT salary
    FROM employees
    WHERE department='HR'
);
```

---

# ALL Operator

Returns TRUE only if the condition matches **all** values.

```sql
SELECT *
FROM employees
WHERE salary > ALL
(
    SELECT salary
    FROM employees
    WHERE department='HR'
);
```

---

# Execution Flow

```text
Inner Query
      │
      ▼
Returns Result
      │
      ▼
Outer Query Executes
```

---

# Complete Example

```sql
-- Average Salary
SELECT AVG(salary)
FROM employees;

-- Salary Greater Than Average
SELECT *
FROM employees
WHERE salary >
(
SELECT AVG(salary)
FROM employees
);

-- Department IN
SELECT *
FROM employees
WHERE department IN
(
SELECT department
FROM employees
WHERE salary > 50000
);
```

### 4. Difference between JOIN and Subquery?

|------|----------|

| Faster for many operations | Easier for nested logic |

# Common Mistakes

❌ Forgetting parentheses.

Wrong

```sql
SELECT *
FROM employees
WHERE salary >
SELECT AVG(salary)
FROM employees;
```

Correct

```sql
SELECT *
FROM employees
WHERE salary >
(
SELECT AVG(salary)
FROM employees
);
```
