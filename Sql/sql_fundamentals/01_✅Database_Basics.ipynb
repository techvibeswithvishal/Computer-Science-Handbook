# SQL Chapter 1: Database Basics



---

# 1. What is a Database?

A Database is an organized collection of data that can be easily stored, managed, and accessed.

### Real-Life Examples

* Student Records
* Bank Accounts
* E-commerce Products
* Employee Information

Example:

| ID | Name   | Age |
| -- | ------ | --- |
| 1  | Vishal | 21  |
| 2  | Rahul  | 22  |

---

# 2. What is DBMS?

DBMS (Database Management System) is software used to create, manage, and manipulate databases.

### Popular DBMS

* MySQL
* PostgreSQL
* Oracle
* SQL Server

---

# 3. What is RDBMS?

RDBMS stands for Relational Database Management System.

Data is stored in the form of tables (Rows and Columns).

Example:

Students Table

| ID | Name   | City  |
| -- | ------ | ----- |
| 1  | Vishal | Delhi |
| 2  | Aman   | Agra  |

---

# 4. What is SQL?

SQL stands for Structured Query Language.

SQL is used to communicate with databases.

Using SQL we can:

* Create Data
* Read Data
* Update Data
* Delete Data

These are called CRUD Operations.

---

# 5. SQL vs MySQL

| SQL                   | MySQL             |
| --------------------- | ----------------- |
| Language              | Database Software |
| Used to write queries | Uses SQL queries  |
| Standard language     | RDBMS             |

Example:

SQL:

```sql
SELECT * FROM students;
```

MySQL:
Software that executes the SQL query.

---

# 6. CRUD Operations

| Operation | SQL Command |
| --------- | ----------- |
| Create    | INSERT      |
| Read      | SELECT      |
| Update    | UPDATE      |
| Delete    | DELETE      |

Example:

```sql
INSERT INTO students VALUES (1,'Vishal',21);

SELECT * FROM students;

UPDATE students
SET age = 22
WHERE id = 1;

DELETE FROM students
WHERE id = 1;
```

---

# 7. Create Database

Syntax:

```sql
CREATE DATABASE college;
```

Example:

```sql
CREATE DATABASE student_management;
```

---

# 8. Delete Database

Syntax:

```sql
DROP DATABASE database_name;
```

Example:

```sql
DROP DATABASE student_management;
```

---

# Quick Revision

✓ Database = Collection of Data

✓ DBMS = Software to Manage Database

✓ RDBMS = Data Stored in Tables

✓ SQL = Language Used to Communicate with Database

✓ CRUD = Create, Read, Update, Delete

✓ CREATE DATABASE = Create New Database

✓ DROP DATABASE = Delete Database
