# 📖 What is a View?

A **View** is a virtual table created from one or more SQL queries.

A View does **not** store data itself. Instead, it stores the SQL query and displays the latest data from the original table whenever it is accessed.

---

# Why Do We Use Views?

Views are used to:

- Simplify complex SQL queries
- Improve security by hiding sensitive columns
- Reuse frequently used queries
- Make SQL easier to read and maintain

---

# Basic Syntax

```sql
CREATE VIEW view_name AS
SELECT column1, column2
FROM table_name;
```

# Create a View

Create a view that displays only employee names and departments.

```sql
CREATE VIEW employee_view AS
SELECT name, department
FROM employees;
```

---

# Display Data from View

```sql
SELECT *
FROM employee_view;
```

Output

| name | department |
|------|------------|
|Vishal|IT|
|Rahul|HR|
|Neha|IT|
|Aman|Finance|

---

# View with WHERE

```sql
CREATE VIEW it_employees AS
SELECT *
FROM employees
WHERE department='IT';
```

Display

```sql
SELECT *
FROM it_employees;
```

---

# Update a View

Some simple views can be updated.

```sql
UPDATE employee_view
SET department='Development'
WHERE name='Vishal';
```

The original table is updated automatically.

---

# Delete a View

```sql
DROP VIEW employee_view;
```

---

# Difference Between Table and View

| Table | View |
|--------|------|
| Stores actual data | Stores only a query |
| Occupies storage | Minimal storage |
| Can exist independently | Depends on table |
| Faster for direct access | May be slower for complex queries |

---

# Advantages of Views

- Improve Security
- Simplify Complex Queries
- Reusable
- Easy Reporting
- Hide Sensitive Columns

---

# Disadvantages of Views

- Depends on the original table
- Complex views may be slower
- Cannot always be updated

---

# Complete Example

```sql
CREATE TABLE employees(
    id INT,
    name VARCHAR(50),
    department VARCHAR(30),
    salary INT
);

INSERT INTO employees VALUES
(1,'Vishal','IT',50000),
(2,'Rahul','HR',40000),
(3,'Neha','IT',60000),
(4,'Aman','Finance',45000);

CREATE VIEW employee_view AS
SELECT name, department
FROM employees;

SELECT * FROM employee_view;
```

# Interview Questions

### 1. What is a View?

A View is a virtual table created from the result of a SQL query.

---

### 2. Does a View store data?

No.

A View stores only the SQL query.

---

### 3. Why are Views used?

- Security
- Simplicity
- Reusability
- Reporting

---

### 4. Can a View be updated?

Yes, simple views can be updated.
Complex views may not be updatable.

---

### 5. Which command deletes a View?

```sql
DROP VIEW view_name;
```

---

# Common Mistakes

❌ Thinking a View stores data.

A View always displays data from the original table.

---

❌ Deleting a View deletes the table.

Wrong.

```text
DROP VIEW

↓

Deletes only the View

NOT the original table.
```
