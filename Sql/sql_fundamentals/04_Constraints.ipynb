# 📘 Chapter 4: Constraints



# 📖 What are Constraints?

Constraints are rules applied to table columns.

They ensure that only valid and accurate data is stored in the database.

Example:

```text
Student Table

ID      → Cannot be duplicate
Name    → Cannot be empty
Age     → Must be a valid number
```

---

# Types of Constraints


```text
Constraints
│
├── PRIMARY KEY
├── FOREIGN KEY
├── NOT NULL
├── UNIQUE
└── DEFAULT
```

---

# 1. PRIMARY KEY

A Primary Key uniquely identifies each record in a table.

Rules:

- Cannot contain NULL values
- Must be unique
- Only one Primary Key per table

### Syntax

```sql
CREATE TABLE students(
    id INT PRIMARY KEY,
    name VARCHAR(50)
);
```

### Example

| id | name |
|----|------|
|1|Vishal|
|2|Rahul|

❌ Not Allowed

| id | name |
|----|------|
|1|Aman|
|1|Neha|

Reason: Duplicate Primary Key

---

# 2. NOT NULL

A NOT NULL column cannot store empty values.

### Syntax

```sql
CREATE TABLE students(
    id INT,
    name VARCHAR(50) NOT NULL
);
```

Allowed

```text
1  Vishal
```

Not Allowed

```text
1  NULL
```

---

# 3. UNIQUE

A UNIQUE column cannot contain duplicate values.

### Syntax

```sql
CREATE TABLE users(
    email VARCHAR(100) UNIQUE
);
```

Allowed

```text
abc@gmail.com

xyz@gmail.com
```

Not Allowed

```text
abc@gmail.com

abc@gmail.com
```

---

# 4. DEFAULT

DEFAULT assigns a value automatically if no value is provided.

### Syntax

```sql
CREATE TABLE employees(
    salary INT DEFAULT 10000
);
```

Example

```sql
INSERT INTO employees
VALUES ();
```

Stored Value

```text
10000
```

---

# 5. FOREIGN KEY

A Foreign Key creates a relationship between two tables.

Example

Students Table

| id | name |
|----|------|
|1|Vishal|
|2|Rahul|

Courses Table

| course_id | student_id |
|-----------|------------|
|101|1|
|102|2|

Here

```text
student_id

↓

references

↓

students.id
```

### Syntax

```sql
CREATE TABLE courses(
    course_id INT PRIMARY KEY,
    student_id INT,
    FOREIGN KEY(student_id)
    REFERENCES students(id)
);
```

---

# Complete Example

```sql
CREATE TABLE students(
    id INT PRIMARY KEY,
    name VARCHAR(50) NOT NULL,
    email VARCHAR(100) UNIQUE,
    city VARCHAR(30) DEFAULT 'Delhi'
);
```

# Interview Questions

### 1. What is a Constraint?

A Constraint is a rule applied to a column to maintain data integrity.

---

### 2. Difference between PRIMARY KEY and UNIQUE?

| PRIMARY KEY | UNIQUE |
|-------------|--------|
| Cannot be NULL | Can contain NULL (MySQL allows one NULL) |
| Must be unique | Must be unique |
| Only one per table | Multiple UNIQUE columns allowed |

---

### 3. Difference between PRIMARY KEY and FOREIGN KEY?

| PRIMARY KEY | FOREIGN KEY |
|-------------|-------------|
| Uniquely identifies a row | Links two tables |
| No duplicates | Duplicate values allowed |
| Parent table | Child table |

---

### 4. Why do we use NOT NULL?

To prevent storing empty values.

---

### 5. Why do we use DEFAULT?

To automatically assign a value when no value is provided.
