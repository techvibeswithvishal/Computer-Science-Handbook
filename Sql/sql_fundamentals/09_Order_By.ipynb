# 📖 What is ORDER BY?

The **ORDER BY** clause is used to sort the result set.

By default, SQL sorts data in **Ascending (ASC)** order.

---

# Basic Syntax

```sql
SELECT column_name
FROM table_name
ORDER BY column_name;
```

---

# Example Table

Students

| id | name | age | city |
|----|------|-----|-------|
|1|Vishal|23|Delhi|

|3|Neha|21|Noida|
|4|Aman|24|Delhi|

---

# 1. ORDER BY ASC (Ascending)

Ascending order is the default order.

Syntax

```sql
SELECT *
FROM students
ORDER BY age ASC;
```

Output

| name | age |
|------|-----|
|Priya|20|
|Neha|21|
|Rahul|22|
|Vishal|23|
|Aman|24|

---

# 2. ORDER BY DESC (Descending)

Displays records from highest to lowest.

Syntax

```sql
SELECT *
FROM students
ORDER BY age DESC;
```

Output

| name | age |
|------|-----|
|Aman|24|
|Vishal|23|
|Rahul|22|
|Neha|21|
|Priya|20|

---

# 3. Sort by Name

```sql
SELECT *
FROM students
ORDER BY name;
```

Output

```text
Aman
Neha
Priya
Rahul
Vishal
```

---

# 4. Sort by Multiple Columns

SQL first sorts by the first column.
If duplicate values exist, it sorts using the second column.

```sql
SELECT *
FROM students
ORDER BY city ASC, age DESC;
```

Example Output

```text
Agra
  Rahul
  Priya

Delhi
  Aman
  Vishal

Noida
  Neha
```

---

# 5. ORDER BY with WHERE

```sql
SELECT *
FROM students
WHERE city='Delhi'
ORDER BY age DESC;
```

Output

| name | age |
|------|-----|
|Aman|24|
|Vishal|23|

---

# SQL Execution Flow

```text
FROM
   ↓
WHERE
   ↓
SELECT
   ↓
ORDER BY
```

---

# Complete Example

```sql
SELECT *
FROM students
ORDER BY age ASC;

SELECT *
FROM students
ORDER BY age DESC;

SELECT *
FROM students
ORDER BY name;

SELECT *
FROM students
ORDER BY city, age DESC;

SELECT *
FROM students
WHERE city='Delhi'
ORDER BY age DESC;
```
