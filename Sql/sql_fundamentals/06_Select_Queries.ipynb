# 📖 What is SELECT?

The SELECT statement is used to retrieve data from a database table.

It is the most frequently used SQL command.

---

# Basic Syntax

```sql
SELECT column_name
FROM table_name;
```

---

# Example Table

Students

| id | name | age | city |
|----|------|-----|-------|
|1|Vishal|23|Delhi|
|2|Rahul|22|Agra|
|3|Neha|21|Noida|

---

# 1. Select All Columns

Syntax

```sql
SELECT * FROM students;
```

Output

| id | name | age | city |
|----|------|-----|-------|
|1|Vishal|23|Delhi|
|2|Rahul|22|Agra|
|3|Neha|21|Noida|

> * means "All Columns".

---

# 2. Select Specific Columns

Syntax

```sql
SELECT name, age
FROM students;
```

Output

| name | age |
|------|-----|
|Vishal|23|
|Rahul|22|
|Neha|21|

---

# 3. Change Column Name (Alias)

Use **AS** to display a different column name.

Syntax

```sql
SELECT column_name AS new_name
FROM table_name;
```

Example

```sql
SELECT name AS Student_Name,
       age AS Student_Age
FROM students;
```

Output

| Student_Name | Student_Age |
|--------------|-------------|
|Vishal|23|
|Rahul|22|

---

# 4. DISTINCT

Used to remove duplicate values.

Example Table

| city |
|------|
|Delhi|
|Agra|
|Delhi|
|Noida|

Query

```sql
SELECT DISTINCT city
FROM students;
```

Output

| city |
|------|
|Delhi|
|Agra|
|Noida|

---

# 5. LIMIT

Used to display a limited number of records.

Syntax

```sql
SELECT *
FROM students
LIMIT 2;
```

Output

| id | name |
|----|------|
|1|Vishal|
|2|Rahul|

---

# Execution Flow

```text
Table
   │
   ▼
SELECT
   │
   ▼
Columns
   │
   ▼
Output
```

---

# Complete Example

```sql
SELECT *
FROM students;

SELECT name, age
FROM students;

SELECT DISTINCT city
FROM students;

SELECT name AS Student_Name
FROM students;

SELECT *
FROM students
LIMIT 2;
```

# Interview Questions

### 1. Which SQL command retrieves data?

```sql
SELECT
```

---

### 2. What does * mean?

It means **All Columns**.

```sql
SELECT * FROM students;
```

---

### 3. What is DISTINCT?

Removes duplicate values.

```sql
SELECT DISTINCT city
FROM students;
```

---

### 4. What is AS?

AS is used to give a temporary name (Alias) to a column.

```sql
SELECT name AS Student_Name
FROM students;
```

---

### 5. What is LIMIT?

LIMIT restricts the number of rows returned.

```sql
SELECT *
FROM students
LIMIT 5;
```

---

# Common Mistakes

❌ Wrong

```sql
SELECT FROM students;
```

No column selected.

---

❌ Wrong

```sql
SELECT name age
FROM students;
```

Missing comma.

Correct

```sql
SELECT name, age
FROM students;
```
