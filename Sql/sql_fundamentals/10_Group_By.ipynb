# 📖 What is GROUP BY?

The **GROUP BY** clause is used to group rows that have the same value in one or more columns.

It is commonly used with **Aggregate Functions** like:

- COUNT()
- SUM()
- AVG()
- MAX()
- MIN()

---

# Basic Syntax

```sql
SELECT column_name, aggregate_function(column_name)
FROM table_name
GROUP BY column_name;
```

# Without GROUP BY

```sql
SELECT COUNT(*)
FROM students;
```

Output

```text
5
```

Returns the total number of students.

---

# With GROUP BY

```sql
SELECT city, COUNT(*)
FROM students
GROUP BY city;
```

Output

| city | COUNT(*) |
|------|-----------|
|Delhi|2|
|Agra|2|
|Noida|1|

Each city becomes one group.

---

# COUNT()

Counts the number of records.

```sql
SELECT city, COUNT(*)
FROM students
GROUP BY city;
```

---

# SUM()

Finds the total.

Example Table

Employees

| name | salary | department |
|------|--------|------------|
|A|30000|IT|
|B|40000|IT|
|C|25000|HR|

```sql
SELECT department, SUM(salary)
FROM employees
GROUP BY department;
```

Output

| department | SUM(salary) |
|------------|-------------|
|IT|70000|
|HR|25000|

---

# AVG()

Finds the average value.

```sql
SELECT city, AVG(age)
FROM students
GROUP BY city;
```

---

# MAX()

Finds the highest value.

```sql
SELECT city, MAX(age)
FROM students
GROUP BY city;
```

---

# MIN()

Finds the lowest value.

```sql
SELECT city, MIN(age)
FROM students
GROUP BY city;
```

---

# GROUP BY Multiple Columns

```sql
SELECT city, age, COUNT(*)
FROM students
GROUP BY city, age;
```

---

# GROUP BY with WHERE

```sql
SELECT city, COUNT(*)
FROM students
WHERE age > 20
GROUP BY city;
```

---

# SQL Execution Order

```text
FROM
   ↓
WHERE
   ↓
GROUP BY
   ↓
SELECT
   ↓
ORDER BY
```

---

# Complete Example

```sql
SELECT city, COUNT(*)
FROM students
GROUP BY city;

SELECT city, AVG(age)
FROM students
GROUP BY city;

SELECT city, MAX(age)
FROM students
GROUP BY city;

SELECT city, MIN(age)
FROM students
GROUP BY city;
```

# Interview Questions

### 1. What is GROUP BY?

GROUP BY groups rows having the same value into one group.

---

### 2. Why is GROUP BY used?

To perform calculations on groups of rows.

---

### 3. Which functions are commonly used with GROUP BY?

```text
COUNT()

SUM()

AVG()

MAX()

MIN()
```

---

### 4. Can GROUP BY be used without Aggregate Functions?

Yes, but it is mainly useful with aggregate functions.

---

### 5. Which clause executes first?

```text
FROM

↓

WHERE

↓

GROUP BY

↓

SELECT

↓

ORDER BY
```

# Common Mistakes

❌ Wrong

```sql
SELECT city, age
FROM students
GROUP BY city;
```

In standard SQL, every selected column must either:

- Be included in GROUP BY
- Be used inside an Aggregate Function

---

Correct

```sql
SELECT city, COUNT(*)
FROM students
GROUP BY city;
```
