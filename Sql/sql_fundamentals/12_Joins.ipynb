# 📖 What is a JOIN?

A JOIN is used to combine data from two or more tables using a related column.

Example

Students Table

| student_id | name |
|------------|------|
|1|Vishal|
|2|Rahul|
|3|Neha|

Courses Table

| course_id | student_id | course |
|-----------|------------|--------|
|101|1|Python|
|102|2|SQL|
|103|1|Java|

Using JOIN, we can display

| name | course |
|------|--------|
|Vishal|Python|
|Rahul|SQL|
|Vishal|Java|

# 1. INNER JOIN

Returns only matching records from both tables.

### Syntax

```sql
SELECT columns
FROM table1
INNER JOIN table2
ON table1.column = table2.column;
```

### Example

```sql
SELECT students.name,
       courses.course
FROM students
INNER JOIN courses
ON students.student_id = courses.student_id;
```

Output

| name | course |
|------|--------|
|Vishal|Python|
|Rahul|SQL|
|Vishal|Java|

---

# 2. LEFT JOIN

Returns all rows from the left table and matching rows from the right table.

If no match exists, NULL is returned.

### Syntax

```sql
SELECT *
FROM students
LEFT JOIN courses
ON students.student_id = courses.student_id;
```

Output

| name | course |
|------|--------|
|Vishal|Python|
|Vishal|Java|
|Rahul|SQL|
|Neha|NULL|

---

# 3. RIGHT JOIN

Returns all rows from the right table and matching rows from the left table.

### Syntax

```sql
SELECT *
FROM students
RIGHT JOIN courses
ON students.student_id = courses.student_id;
```

---

# 4. FULL OUTER JOIN

Returns all matching and non-matching rows from both tables.

> Note:
> MySQL does not support FULL OUTER JOIN directly.
> It is usually achieved using UNION.

---

# 5. CROSS JOIN

Returns every possible combination.

### Example

Students

3 rows

Courses

3 rows

Result

```text
3 × 3 = 9 rows
```

### Syntax

```sql
SELECT *
FROM students
CROSS JOIN courses;
```

---

# 6. SELF JOIN

A table joins with itself.

Used for:

- Employee → Manager
- Parent → Child
- Category → Subcategory

Example

Employees

| emp_id | name | manager_id |
|--------|------|------------|
|1|A|NULL|
|2|B|1|
|3|C|1|

# Complete Example

```sql
SELECT students.name,
       courses.course
FROM students
INNER JOIN courses
ON students.student_id = courses.student_id;

SELECT *
FROM students
LEFT JOIN courses
ON students.student_id = courses.student_id;

SELECT *
FROM students
RIGHT JOIN courses
ON students.student_id = courses.student_id;
```

# Common Mistakes

❌ Forgetting ON condition

Wrong

```sql
SELECT *
FROM students
INNER JOIN courses;
```

Correct

```sql
SELECT *
FROM students
INNER JOIN courses
ON students.student_id = courses.student_id;
```

---

❌ Joining unrelated columns

Always join using related keys.

Example

```text
Students.student_id

↓

Courses.student_id
```
