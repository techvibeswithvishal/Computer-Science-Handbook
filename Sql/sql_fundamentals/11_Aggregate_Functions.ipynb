# 📖 What are Aggregate Functions?

Aggregate Functions perform calculations on multiple rows and return a **single value**.

Example:

```text
Student Marks

80
90
70
95
```

Aggregate Functions can calculate:

- Total Marks
- Average Marks
- Highest Marks
- Lowest Marks
- Number of Students

---

# 1. COUNT()

Returns the total number of rows.

### Syntax

```sql
SELECT COUNT(*)
FROM employees;
```

Output

```text
5
```

Count employees in each department.

```sql
SELECT department, COUNT(*)
FROM employees
GROUP BY department;
```

Output

| Department | Employees |
|------------|-----------|
|IT|3|
|HR|2|

---

# 2. SUM()

Returns the total of a numeric column.

### Syntax

```sql
SELECT SUM(salary)
FROM employees;
```

Output

```text
250000
```

Department-wise Salary

```sql
SELECT department, SUM(salary)
FROM employees
GROUP BY department;
```

---

# 3. AVG()

Returns the average value.

### Syntax

```sql
SELECT AVG(salary)
FROM employees;
```

Department-wise Average Salary

```sql
SELECT department, AVG(salary)
FROM employees
GROUP BY department;
```

---

# 4. MAX()

Returns the highest value.

### Syntax

```sql
SELECT MAX(salary)
FROM employees;
```

Department-wise Maximum Salary

```sql
SELECT department, MAX(salary)
FROM employees
GROUP BY department;
```

---

# 5. MIN()

Returns the lowest value.

### Syntax

```sql
SELECT MIN(salary)
FROM employees;
```

Department-wise Minimum Salary

```sql
SELECT department, MIN(salary)
FROM employees
GROUP BY department;
```

---

# Aggregate Functions with WHERE

Find the total salary of IT employees.

```sql
SELECT SUM(salary)
FROM employees
WHERE department='IT';
```

---

# Aggregate Functions with GROUP BY

Find the average salary of each department.

```sql
SELECT department, AVG(salary)
FROM employees
GROUP BY department;
```

# Common Mistakes

❌ Using SUM() on text columns.

Wrong

```sql
SELECT SUM(name)
FROM employees;
```

SUM() works only with numeric data.

---

❌ Forgetting GROUP BY

Wrong

```sql
SELECT department, AVG(salary)
FROM employees;
```

Correct

```sql
SELECT department, AVG(salary)
FROM employees
GROUP BY department;
```
