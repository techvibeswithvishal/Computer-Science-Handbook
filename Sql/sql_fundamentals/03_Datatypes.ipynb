# Frequently Used SQL Data Types

| Data Type | Description | Usage |
|------------|-------------|-------|
| **CHAR** | Stores a fixed-length string (0–255 characters). | `CHAR(50)` |
| **VARCHAR** | Stores a variable-length string (0–255 characters). | `VARCHAR(50)` |
| **BLOB** | Stores Binary Large Objects (0–65,535 bytes). | `BLOB(1000)` |
| **INT** | Stores integers ranging from **-2,147,483,648** to **2,147,483,647**. | `INT` |
| **TINYINT** | Stores small integers ranging from **-128** to **127**. | `TINYINT` |
| **BIGINT** | Stores large integers ranging from **-9,223,372,036,854,775,808** to **9,223,372,036,854,775,807**. | `BIGINT` |
| **BIT** | Stores bit values. The number of bits can range from **1 to 64**. | `BIT(2)` |
| **FLOAT** | Stores decimal numbers with precision up to **23 digits**. | `FLOAT` |
| **DOUBLE** | Stores decimal numbers with precision from **24 to 53 digits**. | `DOUBLE` |
| **BOOLEAN** | Stores Boolean values (**0** or **1**). | `BOOLEAN` |
| **DATE** | Stores dates in **YYYY-MM-DD** format, ranging from **1000-01-01** to **9999-12-31**. | `DATE` |
| **TIME** | Stores time in **HH:MM:SS** format. | `TIME` |
| **YEAR** | Stores a year in **4-digit format**, ranging from **1901** to **2155**. | `YEAR` |

## Summary

- **Character Types:** `CHAR`, `VARCHAR`
- **Binary Type:** `BLOB`
- **Integer Types:** `TINYINT`, `INT`, `BIGINT`
- **Bit Type:** `BIT`
- **Floating-Point Types:** `FLOAT`, `DOUBLE`
- **Boolean Type:** `BOOLEAN`
- **Date & Time Types:** `DATE`, `TIME`, `YEAR`
