# 📘 Chapter 2: Create Database & Create Table


# Step 1: Create a Database

A database is a collection of related tables.

### Syntax

```sql
CREATE DATABASE database_name;
```

### Example

```sql
CREATE DATABASE college;
```

---

# Step 2: Select the Database

Before creating tables, select the database.

### Syntax

```sql
USE database_name;
```

### Example

```sql
USE college;
```

> **Note:** `USE` is a MySQL command. If you are using SQLite, you don't need this command because connecting to the `.db` file automatically selects the database.

---

# Step 3: Create a Table

A table stores data in rows and columns.

### Syntax

```sql
CREATE TABLE table_name(
    column_name datatype,
    column_name datatype
);
```

### Example

```sql
CREATE TABLE students(
    id INT,
    name VARCHAR(50),
    age INT
);
```

---

# Step 4: View Tables

Display all tables inside the current database.

### Syntax

```sql
SHOW TABLES;
```

---

# Step 5: Describe Table Structure

Display the columns and datatypes of a table.

### Syntax

```sql
DESCRIBE students;
```

or

```sql
DESC students;
```

---

# Step 6: Delete a Table

Remove a table permanently.

### Syntax

```sql
DROP TABLE table_name;
```

### Example

```sql
DROP TABLE students;
```

---

# Step 7: Delete a Database

Remove the database permanently.

### Syntax

```sql
DROP DATABASE database_name;
```

### Example

```sql
DROP DATABASE college;
```

---

# Complete Example

```sql
CREATE DATABASE college;

USE college;

CREATE TABLE students(
    id INT,
    name VARCHAR(50),
    age INT
);

SHOW TABLES;

DESC students;
```
