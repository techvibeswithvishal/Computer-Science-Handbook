# 📖 What are SQL Operators?

SQL Operators are symbols or keywords used to perform operations on data.

They help us:

- Compare values
- Filter records
- Perform calculations
- Search data

# 1. Arithmetic Operators

Used for mathematical calculations.

| Operator | Meaning |
|----------|---------|
| + | Addition |
| - | Subtraction |
| * | Multiplication |
| / | Division |
| % | Modulus |

Example

```sql
SELECT 10 + 5;
```

Output

```text
15
```

Example

```sql
SELECT salary * 12
FROM employees;
```

---

# 2. Comparison Operators

Used to compare values.

| Operator | Meaning |
|----------|---------|
| = | Equal To |
| > | Greater Than |
| < | Less Than |
| >= | Greater Than or Equal To |
| <= | Less Than or Equal To |
| != | Not Equal To |
| <> | Not Equal To |

Example

```sql
SELECT *
FROM students
WHERE age > 22;
```

---

# 3. Logical Operators

Used to combine multiple conditions.

## AND

Both conditions must be TRUE.

```sql
SELECT *
FROM students
WHERE city='Delhi'
AND age>22;
```

---

## OR

At least one condition must be TRUE.

```sql
SELECT *
FROM students
WHERE city='Delhi'
OR city='Agra';
```

---

## NOT

Returns the opposite result.

```sql
SELECT *
FROM students
WHERE NOT city='Delhi';
```

---

# 4. BETWEEN

Used to select values within a range.

Syntax

```sql
SELECT *
FROM students
WHERE age BETWEEN 20 AND 23;
```

---

# 5. IN

Checks whether a value exists in a list.

```sql
SELECT *
FROM students
WHERE city IN ('Delhi','Agra');
```

Equivalent to

```sql
WHERE city='Delhi'
OR city='Agra';
```

---

# 6. NOT IN

```sql
SELECT *
FROM students
WHERE city NOT IN ('Delhi','Agra');
```

---

# 7. LIKE

Used for pattern matching.

## Starts With

```sql
SELECT *
FROM students
WHERE name LIKE 'V%';
```

---

## Ends With

```sql
SELECT *
FROM students
WHERE name LIKE '%a';
```

---

## Contains

```sql
SELECT *
FROM students
WHERE name LIKE '%sh%';
```

---

# Wildcards

| Symbol | Meaning |
|---------|---------|
| % | Any number of characters |
| _ | Exactly one character |

Example

```sql
WHERE name LIKE '_a%';
```

---

# 8. IS NULL

Find NULL values.

```sql
SELECT *
FROM students
WHERE city IS NULL;
```

---

# 9. IS NOT NULL

```sql
SELECT *
FROM students
WHERE city IS NOT NULL;
```

---

# Complete Example

```sql
SELECT *
FROM students
WHERE age BETWEEN 20 AND 23;

SELECT *
FROM students
WHERE city IN ('Delhi','Noida');

SELECT *
FROM students
WHERE city NOT IN ('Delhi');

SELECT *
FROM students
WHERE name LIKE 'A%';

SELECT *
FROM students
WHERE city IS NOT NULL;
```

# Interview Questions

### 1. What are SQL Operators?

Operators are used to compare, calculate, and filter data.

---

### 2. Difference between = and LIKE?

| = | LIKE |
|---|------|
| Exact Match | Pattern Matching |

---

### 3. Difference between IN and BETWEEN?

| IN | BETWEEN |
|----|----------|
| Matches values from a list | Matches values within a range |

---

### 4. Difference between IS NULL and = NULL?

Correct

```sql
WHERE city IS NULL;
```

Incorrect

```sql
WHERE city = NULL;
```

---

### 5. What does % mean in LIKE?

It represents **zero or more characters**.
