# 📘 Chapter 5: INSERT INTO Statement


# 📖 What is INSERT?

The INSERT statement is used to add new records (rows) into a table.

Example:

Student Table

| id | name | age |
|----|------|-----|
| 1 | Vishal | 23 |
| 2 | Rahul | 22 |

---

# INSERT Syntax

```sql
INSERT INTO table_name
VALUES(value1, value2, value3);
```

---

# Example 1: Insert One Row

```sql
CREATE TABLE students(
    id INT,
    name VARCHAR(50),
    age INT
);

INSERT INTO students
VALUES(1,'Vishal',23);
```

---

# Example 2: Insert Multiple Rows

```sql
INSERT INTO students
VALUES
(2,'Rahul',22),
(3,'Aman',24),
(4,'Neha',21);
```

---

# Example 3: Insert Using Column Names

Instead of inserting values for every column, you can specify the column names.

Syntax

```sql
INSERT INTO table_name(column1,column2)
VALUES(value1,value2);
```

Example

```sql
INSERT INTO students(id,name)
VALUES(5,'Priya');
```

Here, only **id** and **name** receive values.

---

# Example 4: Using DEFAULT Value

Table

```sql
CREATE TABLE employees(
    id INT,
    name VARCHAR(50),
    city VARCHAR(20) DEFAULT 'Delhi'
);
```

Insert

```sql
INSERT INTO employees(id,name)
VALUES(1,'Vishal');
```

Output

| id | name | city |
|----|------|-------|
|1|Vishal|Delhi|

---

# Example 5: NULL Value

```sql
INSERT INTO students
VALUES(6,NULL,20);
```

Allowed only if the column is **not** declared as NOT NULL.

---

# Common Mistakes

❌ Number of values does not match columns

```sql
INSERT INTO students
VALUES(1,'Vishal');
```

Error because **age** is missing.

---

❌ Wrong Data Type

```sql
INSERT INTO students
VALUES('ABC','Vishal',23);
```

If **id** is INT, this is invalid.

---

❌ Wrong Column Order

Table

```text
id
name
age
```

Wrong

```sql
INSERT INTO students
VALUES('Vishal',1,23);
```

Correct

```sql
INSERT INTO students
VALUES(1,'Vishal',23);
```

---

# Complete Example

```sql
CREATE TABLE students(
    id INT,
    name VARCHAR(50),
    age INT
);

INSERT INTO students
VALUES
(1,'Vishal',23),
(2,'Rahul',22),
(3,'Neha',21);

SELECT * FROM students;
```

# Interview Questions

### 1. Which statement is used to insert data?

```sql
INSERT INTO
```

---

### 2. Can we insert multiple rows in one query?

Yes.

Example

```sql
INSERT INTO students
VALUES
(1,'A',20),
(2,'B',21),
(3,'C',22);
```

---

### 3. Can we insert values into selected columns?

Yes.

```sql
INSERT INTO students(id,name)
VALUES(1,'Vishal');
```

---

### 4. What happens if the number of values is different from the number of columns?

SQL returns an error.

---

### 5. Can DEFAULT values be inserted automatically?

Yes.

If a DEFAULT constraint exists and the column is omitted, SQL inserts the default value.
