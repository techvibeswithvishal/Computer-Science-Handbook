# 1. What is SQL?

SQL (Structured Query Language) is a language used to create, manage, manipulate, and retrieve data from relational databases.

---

# 2. What is DBMS?

DBMS (Database Management System) is software used to create, store, manage, and retrieve data.

Examples:

- MySQL
- PostgreSQL
- Oracle
- SQL Server

---

# 3. What is RDBMS?

RDBMS (Relational Database Management System) stores data in tables and maintains relationships using keys.

Examples:

- MySQL
- PostgreSQL
- Oracle

---

# 4. Difference between DBMS and RDBMS

| DBMS | RDBMS |
|------|--------|
| Stores data | Stores data in tables |
| No relationships | Supports relationships |
| Less secure | More secure |
| Doesn't always support normalization | Supports normalization |

---

# 5. What is a Primary Key?

A Primary Key uniquely identifies each row.

Properties:

- Unique
- Cannot be NULL
- One Primary Key per table

Example

```sql
CREATE TABLE students(
    id INT PRIMARY KEY,
    name VARCHAR(50)
);
```

---

# 6. What is a Foreign Key?

A Foreign Key connects two tables.

Example

```sql
FOREIGN KEY(student_id)
REFERENCES students(id);
```

---

# 7. Difference between Primary Key and Foreign Key

| Primary Key | Foreign Key |
|-------------|-------------|
| Unique | Duplicate values allowed |
| Cannot be NULL | Can contain NULL |
| Identifies records | Creates relationships |

---

# 8. Difference between CHAR and VARCHAR

| CHAR | VARCHAR |
|------|----------|
| Fixed Length | Variable Length |
| Faster | Saves Storage |
| Stores extra spaces | Stores only actual characters |

---

# 9. Difference between DELETE, TRUNCATE and DROP

| DELETE | TRUNCATE | DROP |
|---------|----------|------|
| Deletes selected rows | Deletes all rows | Deletes entire table |
| WHERE allowed | WHERE not allowed | Table removed |
| Can rollback (in transactions) | Usually faster | Structure also deleted |

---

# 10. Difference between WHERE and HAVING

| WHERE | HAVING |
|--------|---------|
| Filters rows | Filters groups |
| Before GROUP BY | After GROUP BY |
| Doesn't use Aggregate Functions | Used with Aggregate Functions |

---

# 11. Difference between GROUP BY and ORDER BY

| GROUP BY | ORDER BY |
|-----------|----------|
| Groups rows | Sorts rows |
| Used with Aggregate Functions | Used for sorting results |

---

# 12. Difference between INNER JOIN and LEFT JOIN

| INNER JOIN | LEFT JOIN |
|-------------|-----------|
| Only matching rows | All rows from left table + matching rows |

---

# 13. Difference between UNIQUE and PRIMARY KEY

| PRIMARY KEY | UNIQUE |
|-------------|--------|
| Cannot be NULL | Allows one NULL (MySQL) |
| One per table | Multiple allowed |

---

# 14. What are Aggregate Functions?

Aggregate Functions perform calculations on multiple rows.

Examples

```text
COUNT()

SUM()

AVG()

MAX()

MIN()
```

---

# 15. What is a View?

A View is a virtual table created from a SQL query.

It stores the query, not the actual data.

---

# 16. What is a Subquery?

A query inside another SQL query.

Example

```sql
SELECT *
FROM employees
WHERE salary >
(
SELECT AVG(salary)
FROM employees
);
```

---

# 17. What are Constraints?

Constraints enforce rules on table columns.

Examples

```text
PRIMARY KEY

FOREIGN KEY

NOT NULL

UNIQUE

DEFAULT
```

---

# 18. What is Normalization?

Normalization organizes data to reduce redundancy and improve consistency.

Types:

- 1NF
- 2NF
- 3NF

---

# 19. What are SQL Joins?

Joins combine data from multiple tables.

Types

```text
INNER JOIN

LEFT JOIN

RIGHT JOIN

FULL OUTER JOIN

CROSS JOIN

SELF JOIN
```

---

---

# 21. What is NULL?

NULL means a value is missing or unknown.

It is different from:

- 0
- Empty String ('')
- False

---

# 22. What is DISTINCT?

Removes duplicate values.

Example

```sql
SELECT DISTINCT city
FROM students;
```

---

# 23. What is LIMIT?

LIMIT restricts the number of rows returned.

Example

```sql
SELECT *
FROM students
LIMIT 5;
```

---

# 24. What is the Difference Between = and LIKE?

| = | LIKE |
|---|------|
| Exact Match | Pattern Matching |

Example

```sql
WHERE name='Vishal';
```

```sql
WHERE name LIKE 'V%';
```
